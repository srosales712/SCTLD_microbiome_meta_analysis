```python
import pandas as ps
import numpy as np

dir_path='/space/home/rosales/rawdata/2020_SCTLD_16S_meta_analysis/Apprill_USVIApril2019TransmissionExpMucus/LibI_folder/'


header_list = ["filepath"]
files_df=ps.read_table('/space/home/rosales/projects/2020_SCTLD_meta-analysis/files_LibI_Apprill_transmissionMucus2019.txt',
                                    names=header_list)

files_df[['sample-id', "other"]]= files_df.filepath.str.split("_L001",expand=True,)
files_df.loc[files_df['filepath'].str.contains('_R1_'), 'forward-absolute-filepath'] = files_df['filepath'] 
files_df.loc[files_df['filepath'].str.contains('_R1_'), 'reverse-absolute-filepath'] = files_df['filepath']


files_df['forward-absolute-filepath'] = dir_path + files_df['forward-absolute-filepath'].astype(str)
files_df['reverse-absolute-filepath'] = dir_path + files_df['reverse-absolute-filepath'].astype(str)

files_df =files_df[["sample-id","forward-absolute-filepath", "reverse-absolute-filepath"]]

files_df=files_df.drop_duplicates("sample-id")

files_df['reverse-absolute-filepath']=files_df['reverse-absolute-filepath'].str.replace('R1','R2')

files_df.to_csv('/space/home/rosales/projects/2020_SCTLD_meta-analysis/manifest_files/files_Apprill_2019TransmissionExpMucus_LibI.csv',
index=False, header=True, sep ="\t")
```

```python
%%bash
#trim PCR primers
for f1 in /space/home/rosales/projects/2020_SCTLD_meta-analysis/Folder_Apprill_transmissionMucus2019/Import_files_Apprill_2019TransmissionExpMucus_Lib*.qza
do
newfile="$(basename $f1)"
fn=${newfile%*.csv};

qiime cutadapt trim-paired \
--i-demultiplexed-sequences $f1 \
--p-cores 2 \
--p-front-f GTGYCAGCMGCCGCGGTAA \
--p-front-r GGACTACNVGGGTWTCTAAT \
--o-trimmed-sequences Folder_Apprill_transmissionMucus2019/trim_$fn \
--verbose

done

```

```python
#visualize
%%bash

for f1 in Folder_Apprill_transmissionMucus2019/trim_Import_files_Apprill_2019TransmissionExpMucus_Lib*.qza
do
newfile="$(basename $f1)"
fn=${newfile%.qza};

qiime demux summarize  --i-data $f1  --o-visualization Folder_Apprill_transmissionMucus2019/$fn.qzv

done
```

```python
#run DADA2
%%bash

for f1 in Folder_Apprill_transmissionMucus2019/trim_Import_files_Apprill_2019TransmissionExpMucus_Lib*.qza
do
newfile="$(basename $f1)"
fn=${newfile%.qza};

qiime dada2 denoise-single --i-demultiplexed-seqs $f1 --p-max-ee 3 --p-trunc-len 0 --p-n-threads 10 --o-representative-sequences Folder_Apprill_transmissionMucus2019/rep_$fn.qza --o-table Folder_Apprill_transmissionMucus2019/table_$fn.qza --o-denoising-stats Folder_Apprill_transmissionMucus2019/dada_$fn.qza 

done
```
