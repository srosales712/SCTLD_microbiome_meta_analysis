```python
import pandas as pd
import numpy as np
import re
# import file with header "filepath"
header_list = ["filepath"]
files_df=pd.read_table("/space/home/rosales/projects/2020_SCTLD_meta-analysis/files_StewartMedina_FL2019.txt",
                                    names=header_list)
# change from F and R to R1 and R2 to parse on "_R"
files_df['filepath']=files_df['filepath'].str.replace('_F_','_R1_')
files_df['filepath']=files_df['filepath'].str.replace('_R_','_R2_')
#files_df[files_df['filepath'].str.match('PSTR2c-719')]
# parse column filepath on '_R' which creates two columns a sample ID anf the rest of the filename
files_df[['sample-id', "other"]]= files_df.filepath.str.split("_R",expand=True,)
# if string contains R1 in filename make a new coulumn 
files_df.loc[files_df['filepath'].str.contains('_R1_'), 'forward-absolute-filepath'] = files_df['filepath'] 
files_df.loc[files_df['filepath'].str.contains('_R1_'), 'reverse-absolute-filepath'] = files_df['filepath']
#change Reverse and Forward back to the original IDs
files_df['forward-absolute-filepath']=files_df['forward-absolute-filepath'].str.replace('_R1_','_F_')
files_df['reverse-absolute-filepath']=files_df['reverse-absolute-filepath'].str.replace('_R1_','_R_')

files_df['forward-absolute-filepath'] = dir_path + files_df['forward-absolute-filepath'].astype(str)
files_df['reverse-absolute-filepath'] = dir_path + files_df['reverse-absolute-filepath'].astype(str)
# only print three columns needed slected
files_df =files_df[["sample-id","forward-absolute-filepath", "reverse-absolute-filepath"]]
# remove duplicates samples
files_df=files_df.drop_duplicates("sample-id")
name ='StewartMedina_FL2019'
files_df.to_csv('/space/home/rosales/projects/2020_SCTLD_meta-analysis/manifest_files/files_{0}.csv'.format(name),
index=False, header=True, sep ="\t")
```
