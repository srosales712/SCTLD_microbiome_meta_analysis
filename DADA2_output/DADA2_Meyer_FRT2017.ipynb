```python
import pandas as ps
import numpy as np

header_list = ["filepath"]
files_df=ps.read_table('/space/home/rosales/projects/2020_SCTLD_meta-analysis/files_Meyer_FRT2017.txt',
                                    names=header_list)
files_df[['sample-id', "other"]]= files_df.filepath.str.split("_S",expand=True,)
files_df.loc[files_df['filepath'].str.contains('_R1_'), 'forward-absolute-filepath'] = files_df['filepath'] 
files_df.loc[files_df['filepath'].str.contains('_R1_'), 'reverse-absolute-filepath'] = files_df['filepath']
dir_path="/space/home/rosales/rawdata/2020_SCTLD_16S_meta_analysis/Meyer_FRT2017/Meyer_FRT2017_fastqfiles/"
files_df['forward-absolute-filepath'] = dir_path + files_df['forward-absolute-filepath'].astype(str)
files_df['reverse-absolute-filepath'] = dir_path + files_df['reverse-absolute-filepath'].astype(str)
files_df =files_df[["sample-id","forward-absolute-filepath", "reverse-absolute-filepath"]]
files_df=files_df.drop_duplicates("sample-id")
files_df['reverse-absolute-filepath']=files_df['reverse-absolute-filepath'].str.replace('R1','R2')
files_df.to_csv('/space/home/rosales/projects/2020_SCTLD_meta-analysis/manifest_files/files_Meyer_FRT2017.csv',
index=False, header=True, sep ="\t")
```
