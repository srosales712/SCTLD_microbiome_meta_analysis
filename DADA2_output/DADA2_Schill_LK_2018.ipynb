```python
import pandas as pd
import numpy as np
header_list = ["filepath"]
files_df=pd.read_table('/space/home/rosales/projects/2020_SCTLD_meta-analysis/files_Schill_LK_2018.txt',
                                    names=header_list)
files_df[['sample-id', "other"]]= files_df.filepath.str.split("_",expand=True,)
files_df.loc[files_df['filepath'].str.contains('_1.'), 'forward-absolute-filepath'] = files_df['filepath'] 
files_df.loc[files_df['filepath'].str.contains('_1.'), 'reverse-absolute-filepath'] = files_df['filepath']
files_df['forward-absolute-filepath'] = dir_path + files_df['forward-absolute-filepath'].astype(str)
files_df['reverse-absolute-filepath'] = dir_path + files_df['reverse-absolute-filepath'].astype(str)
files_df =files_df[["sample-id","forward-absolute-filepath", "reverse-absolute-filepath"]]
files_df=files_df.drop_duplicates("sample-id")
files_df['reverse-absolute-filepath']=files_df['reverse-absolute-filepath'].str.replace('_1.fastq','_2.fastq')
files_df.to_csv('/space/home/rosales/projects/2020_SCTLD_meta-analysis/manifest_files/files_Schill_LK_2018.csv',
index=False, header=True, sep ="\t")
```
